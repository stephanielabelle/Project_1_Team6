```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
filepath = Path("Resources/Airbnb_Open_Data.csv")
clean_airbnb_df = pd.read_csv(filepath)
clean_airbnb_df.head()
```

```python
clean_airbnb_df['id'].value_counts()
clean_airbnb_df['id'].nunique()
```

```python
id_cleaned_df = clean_airbnb_df.drop_duplicates(subset=['id'],keep='first')
id_cleaned_df['id'].value_counts()
```

```python
id_cleaned_df.count()
```

```python
dropna_df = id_cleaned_df.dropna(subset=['host_identity_verified', 'lat','long','neighbourhood group','instant_bookable','cancellation_policy',
                      'room type','Construction year','price','last review','review rate number','calculated host listings count'])
dropna_df
```

```python
yearday_airbnb_df = dropna_df.loc[dropna_df['availability 365'] <= 365,:]
yearday_airbnb_df
```

```python
yearday_airbnb_df['neighbourhood group'].unique()
```

```python
yearday_airbnb_df['neighbourhood group'].value_counts()
```

```python
yearday_airbnb_df['neighbourhood group']= yearday_airbnb_df['neighbourhood group'].replace({'brookln':'Brooklyn'})
yearday_airbnb_df['neighbourhood group'].value_counts()
```

```python
yearday_airbnb_df['service fee'] = yearday_airbnb_df['service fee'].fillna(0)

yearday_airbnb_df['service fee']
```

```python
yearday_airbnb_df['price'] = yearday_airbnb_df['price'].str.replace("$","")
yearday_airbnb_df['price'] = yearday_airbnb_df['price'].str.replace(",","")
yearday_airbnb_df['price'] = pd.to_numeric(yearday_airbnb_df['price'])
yearday_airbnb_df['price']
```

```python
yearday_airbnb_df['service fee'] = yearday_airbnb_df['service fee'].str.replace("$","")
yearday_airbnb_df['service fee'] = yearday_airbnb_df['service fee'].str.replace(",","")
yearday_airbnb_df['service fee'] = pd.to_numeric(yearday_airbnb_df['service fee'])
```

```python
yearday_airbnb_df.to_csv("Resources/cleaned_airbnb_data.csv", index=False)
```
