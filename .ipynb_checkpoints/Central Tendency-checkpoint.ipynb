# Central tendency

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

from pathlib import Path
```

```python
filepath = Path("Resources/cleaned_airbnb_data.csv")
clean_df = pd.read_csv(filepath, low_memory=False)
clean_df.head()
```

```python
clean_df['id'].nunique()
```

```python
clean_df.dtypes
```

```python
clean_df['neighbourhood group'].value_counts()
```

```python
clean_df['price_with_currency'] = '$' + clean_df['price'].astype(str)
clean_df['price_with_currency']
```

```python
cleaned_df = clean_df.groupby(["neighbourhood group"])
cleaned_df.head()
```

```python
#Mean,Median,Min & Max for neighbourhood group by city
group_df = clean_df.groupby('neighbourhood group').agg({
    'price': ['mean','median','min','max'],
}).reset_index()
group_df.head()
```

```python
#Pie chart for Mean
pies = ["Bronx", "Brooklyn", "Manhattan", "Queens", "Staten Island"]
pie_votes = ["630.53462", "627.654762", "624.346316", "630.139808", "623.234356"]
colors = ["yellow","green","lightblue","orange","lightpink"]
explode = (0.1,0,0,0,0)
```

```python
# Tell matplotlib to create a pie chart based upon the above data
plt.pie(pie_votes, explode=explode, labels=pies, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Create axes which are equal so we have a perfect circle
plt.axis("equal")
# Save an image of our chart and print the final product to the screen
plt.savefig("Resources/meanpie.png")
plt.show()
```

```python
#Pie chart for Median
pies = ["Bronx", "Brooklyn", "Manhattan", "Queens", "Staten Island"]
pie_votes = ["643.0", "627.0", "624.0", "630.0", "625.0	"]
colors = ["yellow","green","lightblue","orange","lightpink"]
explode = (0.1,0,0,0,0)
```

```python
# Tell matplotlib to create a pie chart based upon the above data
plt.pie(pie_votes, explode=explode, labels=pies, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Create axes which are equal so we have a perfect circle
plt.axis("equal")
# Save an image of our chart and print the final product to the screen
plt.savefig("Resources/medianpie.png")
plt.show()
```

```python
# Filter the DataFrame down only to those columns to chart
central_tendency = group_df[["neighbourhood group","price"]]

# Set the index to be "neighbourhood group" so they will be used as labels
central_tendency = central_tendency.set_index("neighbourhood group")

central_tendency.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
central_tendency.plot(kind="bar", figsize=(20,3.5))

# Set a title for the chart
plt.title("Central Tendency")

plt.savefig("Resources/centraltendency.png")

plt.show()
plt.tight_layout()
```

```python
# Determine which measure of central tendency is most appropriate to describe the Population
plt.hist(central_tendency['price'])
plt.xlabel('Price')
plt.ylabel('Group')
plt.show()
print(central_tendency['price'].mean())
print(central_tendency['price'].median())
```
